#### Description of the features 
**Country**:
   - This column represents the name of the country in which the data was recorded. It likely corresponds to the geographic location where the COVID-19 statistics were gathered.

**Country_code**:

**Month**:

**Agegroup**:
   - This column may categorize the population by age group, such as "0-14", "15-64", and "65+" years. This data is often used to analyze how different age groups are affected by the disease.

**Deaths**:
   - This column represents the number of deaths attributed to COVID-19 in a given country, during the specified year and month, and for a specific age group (if relevant). It helps assess the severity and impact of the pandemic in each country.

#### 4. Checking and Handling Missing Values

```python
# Count missing values
missing = df.isnull().sum()
print("Missing values per column:\n", missing)

# Visualize missing data
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values Heatmap")
plt.show()
```

#### 5. Handle Missing Values
Letâ€™s specifically handle the 'Deaths' column which contains NaN.

```python
# Fill with 0 (if NaN means no reported deaths)
df['Deaths'] = df['Deaths'].fillna(0)
```

```python
# Replace null values in features 'Who_region' and 'Wb_income'
df[['Who_region', 'Wb_income']] = df[['Who_region', 'Wb_income']].fillna("unknown")
```

```python
missing = df.isnull().sum()
print(missing)
```

#### 8. Basic Visualizations
##### A. Deaths Over Time

```python
# Create a 'Date' column by combining 'Year' and 'Month' (assuming the 'Year' and 'Month' columns are integers)
df['Date'] = pd.to_datetime(df[['Year', 'Month']].assign(DAY=1))

# Group by 'Date' and sum 'Deaths'
df_grouped = df.groupby('Date')['Deaths'].sum().reset_index()

# Plotting
plt.figure(figsize=(14, 6))
sns.lineplot(data=df_grouped, x='Date', y='Deaths')
plt.title("Total Deaths Over Time")
plt.xlabel("Date")
plt.ylabel("Deaths")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
region_group = df.groupby('Who_region')['Deaths'].sum().sort_values(ascending=False)
sns.barplot(x=region_group.index, y=region_group.values)
plt.title("Total Deaths by WHO Region")
plt.ylabel("Deaths")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##### Top 5 Countries in Each Year Based on Deaths

```python
# Group by Year and Country, summing total deaths
year_country_deaths = df.groupby(['Year', 'Country'])['Deaths'].sum().reset_index()

# For each year, get the top 5 countries with the highest deaths
top5_per_year = year_country_deaths.groupby('Year').apply(
    lambda x: x.nlargest(5, 'Deaths')
).reset_index(drop=True)

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(data=top5_per_year, x='Year', y='Deaths', hue='Country')
plt.title('Top 5 Countries by Deaths per Year')
plt.xlabel('Year')
plt.ylabel('Total Deaths')
plt.xticks(rotation=45)
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

* Widespread vaccination,

* Better awareness and control measures.

iii. High Death Toll in 2020â€“2021 for Latin American Countries Countries such as Brazil, Mexico, and Peru appear prominently in the top 5 during 2020 and 2021, reflecting how severely they were affected in the earlier phase of the pandemic.

iv. Emergence of European Countries in Later Years After 2022, European countries like Italy, Germany, France, and Spain start to appear in the top 5, although with relatively lower death counts than earlier leaders.

#### 9. Encoding Categorical Variables

```python
# Remove columns which is not relevant to the analysis, for example, Country_code
df=df.drop(columns=['Date', 'Country_code'])
```


* Encoding categorical variables is a critical step in data preprocessing for machine learning models, as most models require numerical input.
##### There are two approach:
* **Label encoding** and **one-hot encoding.**
**Label Encoding**
* It converts each category in a categorical variable to a unique integer.
  
**When to Use:** When the categorical variable has an ordinal relationship (e.g., low, medium, high).
* When there are a limited number of categories.
  
**One-Hot Encoding**
* One-hot encoding converts categorical variables into a series of binary columns, each representing a unique category.
  
**When to Use**

* When the categorical variable is nominal (no intrinsic order).
* When you want to avoid introducing ordinal relationships.

```python
from sklearn.preprocessing import LabelEncoder

# One-hot encode categorical columns: 'Country' and 'Who_region'
df_encoded = pd.get_dummies(df, columns=['Country', 'Who_region'], drop_first=True)

# Initialize LabelEncoder
le = LabelEncoder()

# Apply label encoding to 'Wb_income' and 'Agegroup'
df_encoded['Wb_income'] = le.fit_transform(df_encoded['Wb_income'].replace('unknown', '0'))  # Replace 'unknown' with '0' to avoid errors
df_encoded['Agegroup'] = le.fit_transform(df_encoded['Agegroup'])

# Check the result
df_encoded.info()
```
